```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
import pandas as pd
dict_users = {'Drive': 85, 'Public Transport': 5, 'Home': 6, 'Walk': 3, 'Other': 1}
df = pd.Series(dict_users)
```

```python
num_rows = 10
row = 0
column = 0
data = []
colors_list = ['red', 'limegreen', 'darkblue', 'black', 'grey']
colors = {df.index[i]:colors_list[i] for i in range(len(df))}
for category, count in df.iteritems():
    rows = []
    columns = []
    for _ in range(count):
        if row == num_rows:
            row = 0
            column += 1
        rows.append(row)
        columns.append(column)        
        row += 1
```

```python
ig = plt.figure(FigureClass=Waffle,
                 figsize=(10,5),
                 values=dict_users,
                 rows=10,
                 colors=list(colors.values()),
                 icons=['car','bus', 'home', 'walking', 'question'],
                 font_size=22,
                 icon_legend=True,
                 legend={'bbox_to_anchor': (1.55, 1), 'fontsize': 15, 'frameon': False})
plt.title('Means of Transportation to Work in the United States (Rounded)', fontsize=25)
plt.show()
```
