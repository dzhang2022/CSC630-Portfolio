```python
import altair as alt
import pandas as pd
from vega_datasets import data

# Since the data is more than 5,000 rows we'll import it from a URL
source = data.seattle_temps.url
df = pd.read_csv('30 Day Challenge Data/mass_vaccines.csv')

df.head()
```

```python
alt.Chart(
    df,
    title="Daily Vaccinations in Massachusetts"
).mark_rect().encode(
    x='day:O',
    y=alt.Y('month:N', sort=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']),
    color=alt.Color('daily_vaccinations:Q', scale=alt.Scale(scheme="inferno")),
    tooltip=[
        alt.Tooltip('date', title='Date'),
        alt.Tooltip('daily_vaccinations:Q', title='Vaccinations')
    ]
).properties(width=550)
```
