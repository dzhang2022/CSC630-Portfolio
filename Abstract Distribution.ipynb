```python
import numpy as np
import matplotlib.pyplot as plt
import math

dist = np.arange(-5, 5, 0.1)

speed = []
for i in range(100):
    if i > 35 and i < 65:
        speed.append(10)
    else:
        speed.append(1)

speeds = np.array(speed)

with plt.xkcd():
    figure, axes = plt.subplots()
    plt.plot(dist, speeds)
    axes.set_aspect(1)
    axes.add_artist(plt.Circle((0.9, 5), 0.2))
    plt.title('My Walking Speed at Midnight in Relation to the Distance to my Basement Door')
    plt.axhline(y = 0, color ='k')
    plt.xlabel('Distance')
    plt.ylabel('Speed')
     
    plt.show()
```
