```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('population.csv')
df['Year'].shape[0]
df
```

```python
USApop, DEUpop, CHNpop, JPNpop, KORpop = [], [], [], [], []
for x in range(df.shape[0]):
    #USA's Population
    if df['Country Code'][x] == 'USA':
        USApop.append(df['Value'][x])
    
    #Germany's Population
    if df['Country Code'][x] == 'DEU':
        DEUpop.append(df['Value'][x])
    
    #China's Population
    if df['Country Code'][x] == 'CHN':
        CHNpop.append(df['Value'][x])
    
    #Japan's Population
    if df['Country Code'][x] == 'JPN':
        JPNpop.append(df['Value'][x])

    #South Korea's Population
    if df['Country Code'][x] == 'KOR':
        KORpop.append(df['Value'][x])
```

```python
Years = []
for x in range(1960, 2019):
    Years.append(x)
print(Years)
```

```python
from plotnine import ggplot, aes, geom_line, labs
```

```python
(
    
    ggplot()
    + aes(x = 'Years', y = 'USApop')
    + geom_line()
    + labs(title = 'Population of the USA from 1960 to 2018', x = 'Year', y = 'USA Population')
    
)
```

```python
(
    
    ggplot()
    + aes(x = 'Years', y = 'DEUpop')
    + geom_line()
    + labs(title = 'Population of Germany from 1960 to 2018', x = 'Year', y = 'Germany Population')
    
)
```
